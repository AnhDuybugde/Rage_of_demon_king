```python
import pandas as pd

# Đọc file CSV
df = pd.read_csv("Data/SS8_tour1.csv")  # cột: 'player', 'power'

# Sắp xếp theo power giảm dần
df_sorted = df.sort_values(by='power', ascending=False).reset_index(drop=True)
```

```python
for idx, player in enumerate(df_sorted['player'], 1):
    line = f"{idx}. {player}\n"
    print(line.strip())
```

```python
with open("Data/sorted_player.txt", "w", encoding="utf-8") as f:
    # for idx, match in enumerate(matches, 1):
    #     line = f"Trận {idx}: {match[0]} (Power: {match[1]}) vs {match[2]} (Power: {match[3]})\n"
    for idx, player in enumerate(df_sorted['player'], 1):
        line = f"{idx},{player}\n"
        f.write(line)
```
