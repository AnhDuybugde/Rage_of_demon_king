```python
import pandas as pd

# Đọc file CSV
df = pd.read_csv("Data/SS8_zenith1.csv")  # cột: 'player', 'power'

# Sắp xếp theo power giảm dần
df_sorted = df.sort_values(by='power', ascending=False).reset_index(drop=True)
```

```python
with open("Data/sorted_player.txt", "w", encoding="utf-8") as f:
    for idx, row in df_sorted.iterrows():
        line = f"{idx+1},{row['player']},{row['power']}\n"
        f.write(line)
```
